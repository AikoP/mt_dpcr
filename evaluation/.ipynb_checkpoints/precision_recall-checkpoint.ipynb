```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

import utils.readOBJ as reader
import utils.dpcr_utils as utils
import utils.dpcr_generator as generator
from models import detector_mish, detector_relu, detector_swish

import torch
import torch.nn as nn
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import time

device = torch.device("cpu")

if torch.device("cuda"):
    device = torch.device("cuda")
    print("Using", torch.cuda.device_count(), "CUDA devices")
    
torch.no_grad()

seed_file = open('../utils/seed.txt', "r")
seed = int(seed_file.read())
seed_file.close()

print ("Using Seed:", seed)
    
torch.manual_seed(seed)
np.random.seed(seed)
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'Century Schoolbook'
rcParams['font.cursive'] = 'Century Schoolbook'
rcParams['font.size'] = '19'

rcParams['text.usetex'] = True 
rcParams['text.latex.preamble'] = [r'\usepackage{fouriernc}']
```

```python
model_type = 'mish'
k = 20
```

```python
if model_type == 'mish':
    detector = detector_mish.Model(d = 3, k = k, emb_dims = 1024).to(device)
elif model_type == 'swish':
    detector = detector_swish.Model(d = 3, k = k, emb_dims = 1024).to(device)
elif model_type == 'relu':
    detector = detector_relu.Model(d = 3, k = k, emb_dims = 1024).to(device)

detector = nn.DataParallel(detector)

detector_checkpoint = torch.load('../hpc/results/detector_mish_radam_sd/2020-11-09_052228/detector_checkpoints.t7')
detector.load_state_dict(detector_checkpoint['model_state_dict'][-1])
```

```python
start = time.time()

val_pts, _ = reader.readOBJ('../data/models/faces/clean/face_07_clean.obj')
val_pts = np.array(val_pts).astype(float)

pts_tensor = torch.from_numpy(val_pts).float().to(device)

# mean centering
pts_tensor = pts_tensor - torch.mean(pts_tensor, dim = 0)

min_vals, _ = torch.min(pts_tensor, dim = 0)
max_vals, _ = torch.max(pts_tensor, dim = 0)
scale = torch.max(torch.abs(min_vals),  torch.abs(max_vals))

# scale to fit 2x2x2 bounding box
pts = pts_tensor / scale

knn = utils.knn(pts, 6)

val_data = generator.getData(pts, 200)

neighbors_dirs = pts[knn] - pts[:, None, :]

print ("Generated %d Test Models" % len(val_data))
print ("Total Time:", time.time() - start)
```

```python
start = time.time()

val_predictions = []

for val_sample in val_data:

    pts_ids = val_sample[:,0]

    edge_mask_np = val_sample[:,1].clone().detach().cpu().numpy()
    sample = pts_tensor[pts_ids]

    val_predictions.append(detector(sample.unsqueeze(0).transpose(1,2).to(device)).squeeze(0).transpose(0,1).clone().detach().cpu().numpy())
    
print ("Total Time:", time.time() - start)
```

```python
def getAverageAccuracy(val_data, val_predictions, thresholds = np.array([0.5])):
    
    cum_accuracy = np.zeros(thresholds.shape[0])
    cum_c0_accuracy = np.zeros(thresholds.shape[0])
    cum_c1_accuracy = np.zeros(thresholds.shape[0])
    
    for (i, val_sample) in enumerate(val_data):
        
        edge_mask_np = val_sample[:,1].clone().detach().cpu().numpy()
        
        prediction_softmax = (np.exp(val_predictions[i]) / np.sum(np.exp(val_predictions[i]), axis=1)[:,None])
        prediction_softmax = prediction_softmax[:,1] - prediction_softmax[:,0]
        
        prediction = prediction_softmax[:, None] > thresholds[None, :]
        
        cum_accuracy += np.sum(prediction == edge_mask_np[:, None], axis = 0) / edge_mask_np.shape[0]
        cum_c0_accuracy += np.sum(prediction[edge_mask_np == 0] == 0, axis = 0) / edge_mask_np[edge_mask_np == 0].shape[0]
        cum_c1_accuracy += np.sum(prediction[edge_mask_np == 1] == 1, axis = 0) / edge_mask_np[edge_mask_np == 1].shape[0]
        
    return cum_accuracy / len(val_data), cum_c0_accuracy / len(val_data), cum_c1_accuracy / len(val_data)
```

```python
acc, acc_c0, acc_c1 = getAverageAccuracy(val_data, val_predictions, thresholds = np.array([0.0001]))

print ("Average Accuracy:", acc)
print ("Average C0 Accuracy:", acc_c0)
print ("Average C1 Accuracy:", acc_c1)
```

```python
T = np.linspace(-1,1,100)
acc, acc_c0, acc_c1 = getAverageAccuracy(val_data, val_predictions, thresholds = T)
```
