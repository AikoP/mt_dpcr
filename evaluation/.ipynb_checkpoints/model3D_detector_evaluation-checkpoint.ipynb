### Imports & Preparation

```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

from models3D import detector3D, predictor3D
from utils import dpcr_utils, dpcr_generator

import torch
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import time
from mpl_toolkits.mplot3d import Axes3D

device = torch.device("cpu")

if torch.device("cuda"):
    device = torch.device("cuda")
    print("Using", torch.cuda.device_count(), "CUDA devices")

seed_file = open('../utils/seed.txt', "r")
seed = int(seed_file.read())
seed_file.close()

print ("Using Seed:", seed)
    
torch.manual_seed(seed)
np.random.seed(seed)
```

### Checkpoint Loading

```python
detector = detector3D.Model(k = 10, emb_dims = 1024, dropout = 0.5).to(device)
detector = torch.nn.DataParallel(detector)

predictor = predictor3D.Model(k = 20, emb_dims = 1024, dropout = 0.5).to(device)
predictor = torch.nn.DataParallel(predictor)

detector_checkpoint = torch.load('../models3D/model_detector_3d_batched_e50.t7')
predictor_checkpoint = torch.load('../models3D/model_predictor_3d_batched_e70.t7')

detector.load_state_dict(detector_checkpoint['model_state_dict'])
print ("> Loaded detector model (%d epochs)" % detector_checkpoint['epoch'])

predictor.load_state_dict(predictor_checkpoint['model_state_dict'][-1])
print ("> Loaded predictor model (%d epochs)" % len(predictor_checkpoint['train_time']))

_ = detector.eval()
_ = predictor.eval()
```

### Test Model Generation

```python
start = time.time()

model_file = open('../data/models/bunny_5K.obj', 'r') 
Lines = model_file.readlines() 
model_file.close()
  
model = []

for line in Lines: 
    
    ln_arr = line.strip().split(" ")
    if len(ln_arr) == 4 and ln_arr[0] == 'v':
        model.append(ln_arr[1:])
        
model = np.array(model).astype(float)

testArr = dpcr_generator.getTrainingArrayFromModelThreaded(model, 100, 20, gamma = 1.1)

print ("Generated %d Test Models" % (len(testArr)))
print ("Total Time:", time.time() - start)
```

```python
fig = plt.figure(figsize=(16,16))
ax = Axes3D(fig)

sample = testArr[np.random.randint(len(testArr))]

print (sample.shape)

edge_pts = sample[sample[:, 6] == True, 0:3]
non_edge_pts = sample[sample[:, 6] == False, 0:3]

ax.scatter(non_edge_pts[:,0], non_edge_pts[:,1], non_edge_pts[:,2])
ax.scatter(edge_pts[:,0], edge_pts[:,1], edge_pts[:,2])

plt.show()
```
