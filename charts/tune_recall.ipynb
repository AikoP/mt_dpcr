```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

from utils import reader
from utils import dpcrUtils as utils
from utils import generator

from models import models

import torch
import torch.nn as nn
import numpy as np
import scipy as sp
import matplotlib
import matplotlib.pyplot as plt
import time

import torch.nn as nn
pooler = nn.AvgPool1d(1, 1)

device = torch.device("cpu")

if torch.device("cuda"):
    device = torch.device("cuda")
    print("Using", torch.cuda.device_count(), "CUDA devices")
    
torch.no_grad()

seed_file = open('../utils/seed.txt', "r")
seed = int(seed_file.read())
seed_file.close()

print ("Using Seed:", seed)
    
torch.manual_seed(seed)
np.random.seed(seed)
```

```python
start = time.time()

val_pts, _ = reader.readOBJ('../data/single_model_training/bunny/bunny.obj')
val_pts = np.array(val_pts).astype(float)

pts = torch.from_numpy(val_pts).float().to(device)

# mean centering & scaling
pts = pts - torch.mean(pts, dim = 0)
pts = pts / torch.max(torch.abs(pts))

knn = utils.knn(pts, 6)

hidden_range = (6,11)
iteration_range = (6,9)

val_data = generator.getSampleList(knn, 250, iteration_range = iteration_range, hidden_range=hidden_range)

neighbors_dirs = pts[knn] - pts[:, None, :]

torch.cuda.empty_cache()

print ("Generated %d Test Models" % len(val_data))
print ("Total Time:", time.time() - start)
```

```python
def getAverageAccuracy(val_data, val_predictions, thresholds = np.array([0.5]), raw_data=True):
    
    accuracies = np.zeros((len(val_data), thresholds.shape[0]))
    c0_accuracies = np.zeros((len(val_data), thresholds.shape[0]))
    c1_accuracies = np.zeros((len(val_data), thresholds.shape[0]))
    
    for (i, val_sample) in enumerate(val_data):
        
        edge_mask_np = val_sample[:,1].clone().detach().cpu().numpy()
        
        prediction_softmax = np.exp(val_predictions[i]) / np.sum(np.exp(val_predictions[i]), axis=1)[:,None]
        prediction_softmax = prediction_softmax[:,1] - prediction_softmax[:,0]
        
        prediction = prediction_softmax[:, None] > thresholds[None, :]
        
        accuracies[i] = np.sum(prediction == edge_mask_np[:, None], axis = 0) / edge_mask_np.shape[0]
        c0_accuracies[i] += np.sum(prediction[edge_mask_np == 0] == 0, axis = 0) / edge_mask_np[edge_mask_np == 0].shape[0]
        c1_accuracies[i] += np.sum(prediction[edge_mask_np == 1] == 1, axis = 0) / edge_mask_np[edge_mask_np == 1].shape[0]
        
    if raw_data:
        return accuracies, c0_accuracies, c1_accuracies
    else:
        return np.mean(accuracies, axis=0), np.mean(c0_accuracies, axis=0), np.mean(c1_accuracies, axis=0)
```

```python
checkpoint_detector = torch.load('../hpc/results/detector_single_bunny_cnet_mish_radam_sd/2020-12-08_110946/detector_checkpoints.t7')

detector = models.getModel(checkpoint_detector['train_settings'][-1]['model_args']).to(device)

T = np.linspace(-1,1,100)

acc_c0_means = []
acc_c1_means = []
acc_c0_stds = []
acc_c1_stds = []

for i in range(len(checkpoint_detector['model_state_dict'])):
    
    #detector.load_state_dict(checkpoint_detector['model_state_dict'][i])
    state_dict = checkpoint_detector['model_state_dict'][i]
    state_dict['base'] = torch.empty(0).to(device)
    detector.load_state_dict(state_dict)
    detector.eval()
    
    with torch.no_grad():

        torch.cuda.empty_cache()

        start = time.time()

        val_predictions = []

        for val_sample in val_data:

            if val_sample is None:
                print ("None!?")
                continue

            pts_ids = val_sample[:,0]

            edge_mask_np = val_sample[:,1].clone().detach().cpu().numpy()
            sample = pts[pts_ids]

            val_predictions.append(detector(sample.unsqueeze(0).transpose(1,2).to(device)).squeeze(0).transpose(0,1).clone().detach().cpu().numpy())

    
    acc, acc_c0, acc_c1 = getAverageAccuracy(val_data, val_predictions, thresholds = T)
    
    acc_samples = pooler(torch.from_numpy(acc).transpose(0,1).unsqueeze(0)).squeeze(0).transpose(0,1).cpu().numpy()
    acc_c0_samples = pooler(torch.from_numpy(acc_c0).transpose(0,1).unsqueeze(0)).squeeze(0).transpose(0,1).cpu().numpy()
    acc_c1_samples = pooler(torch.from_numpy(acc_c1).transpose(0,1).unsqueeze(0)).squeeze(0).transpose(0,1).cpu().numpy()

    acc_mean = np.mean(acc_samples, axis=0)
    acc_c0_mean = np.mean(acc_c0_samples, axis=0)
    acc_c1_mean = np.mean(acc_c1_samples, axis=0)

    acc_std = np.std(acc_samples, axis=0)
    acc_c0_std = np.std(acc_c0_samples, axis=0)
    acc_c1_std = np.std(acc_c1_samples, axis=0)
    
    acc_c0_means.append(acc_c0_mean)
    acc_c1_means.append(acc_c1_mean)
    acc_c0_stds.append(acc_c0_std)
    acc_c1_stds.append(acc_c1_std)
    
    print ("Total Time:", time.time() - start)
```

```python
def plotAccuracies(acc1, acc2, acc1_std, acc2_std, save_path = None):

    fig = plt.figure(figsize=(16,13))

    ax = plt.axes()

    plt.setp(ax.spines.values(), color=3 * [0.5])
    ax.set_facecolor(3 * [0.99])
    ax.tick_params(axis='x', colors=3 * [0.3])
    ax.tick_params(axis='y', colors=3 * [0.3])

    for (i,(acc1_mean, acc2_mean)) in enumerate(zip (acc1, acc2)):
        color = matplotlib.cm.get_cmap('viridis')((i+1) / len(acc1))
        ax.plot(acc1_mean, acc2_mean, label='iter %d' % i, color = color)
        #ax.plot(acc1_mean - acc1_std[i], acc2_mean - acc2_std[i], label='iter %d' % i, color = color, linestyle=':')
        ax.scatter(0.5 * (acc1_mean[49] + acc1_mean[50]), 0.5 * (acc2_mean[49] + acc2_mean[50]), color = color)
        
    ax.set_xlabel(r'$C_0$ Recall', labelpad=20)
    ax.set_ylabel(r'$C_1$ Recall', labelpad=20)

    ax.set_xlim(0.945,1)
    ax.set_ylim(0.945,1)

    ax.axhline(y = 0.95, color='grey', alpha=0.5, linestyle='--')
    ax.axvline(x = 0.95, color='grey', alpha=0.5, linestyle='--')
    ax.plot([0,1], [0,1], color='grey', alpha=0.3, linestyle=':')

    mappable = matplotlib.cm.ScalarMappable(norm=None, cmap=matplotlib.cm.get_cmap('viridis', len(acc1)))
    mappable.set_array(1+np.arange(len(acc1)))

    cbar = fig.colorbar(mappable, ax=ax, drawedges=True)
    tick_locs = (np.arange(len(acc1)+1) + 0.5)*(len(acc1)-1)/len(acc1)
    cbar.set_ticks(tick_locs)
    cbar.set_ticklabels(1+np.arange(len(acc1)))

    ax.set_axisbelow(True)
    ax.grid(color=3 * [0.88])

    if save_path != None:
        plt.savefig(save_path)
```

```python
plotAccuracies(acc_c0_means[:15], acc_c1_means[:15], acc_c0_stds[:15], acc_c1_stds[:15])
plt.savefig('detector_tradeoff_training.pdf')
```

```python
train_data = None
with open('../data/single_model_training/simple_shapes/single_cube_train_data', 'rb') as file:
    train_data = torch.load(file)

val_pts = train_data["val_pts"].to(device)
val_samples = train_data["val_samples"]
```

```python
from torch.nn.functional import cross_entropy

cum_loss = 0.0
cum_acc = 0.0
cum_acc_c0 = 0.0
cum_acc_c1 = 0.0

class_weights = torch.tensor([0.0253, 0.9747]).to(device)

checkpoint_detector = torch.load('../training/checkpoints/detector_single_cube_cnet_plus_mish_radam_sd/2020-11-27_180404/detector_checkpoints.t7')

detector = models.getModel(checkpoint_detector['train_settings'][-1]['model_args'])

state_dict = checkpoint_detector['model_state_dict'][-2]
new_dict = {}

for key in state_dict:
    new_dict['module.' + key] = state_dict[key]
    
detector.load_state_dict(new_dict)
detector.eval()

for sample in val_samples:

    sample_ids = sample[:,0].flatten().unsqueeze(0)     # size: (1, N_sample)

    # inp = pts[sample_ids].transpose(1,2).contiguous()     # size: (1, 3, N_sample)
    inp = val_pts[sample_ids].transpose(1,2)                    # size: (1, 3, N_sample)

    # p = model(inp).transpose(1,2).squeeze(0).contiguous()     # size: (N_sample, 2)
    p = detector(inp).transpose(1,2).squeeze(0)                    # size: (N_sample, 2)
    t = sample[:,1].flatten().to(p.device)                      # size: (N_sample)

    cum_loss = cum_loss + cross_entropy(p, t, class_weights, reduction='mean').item()

    success_vector = torch.argmax(p, dim=1) == t

    cum_acc += torch.sum(success_vector).item() / success_vector.numel()
    cum_acc_c0 += torch.sum(success_vector[t == 0]).item() / torch.sum(t == 0).item()
    cum_acc_c1 += torch.sum(success_vector[t == 1]).item() / torch.sum(t == 1).item()

#cum_loss/len(samples), cum_acc/len(samples), cum_acc_c0/len(samples), cum_acc_c1/len(samples)
```
