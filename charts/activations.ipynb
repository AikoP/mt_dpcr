```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

import utils.dpcr_utils as utils

import numpy as np
import scipy as sp
import scipy.stats
import scipy.optimize
import scipy.misc
import matplotlib
import matplotlib.pyplot as plt
import scipy.spatial
import time
import itertools
import torch

device = torch.device("cuda")

print('Using GPU : ' + str(torch.cuda.current_device()) + ' from ' + str(torch.cuda.device_count()) + ' devices')

from matplotlib import rcParams
rcParams['font.family'] = 'Century Schoolbook'
rcParams['font.cursive'] = 'Century Schoolbook'
rcParams['font.size'] = '19'

rcParams['text.usetex'] = True 
rcParams['text.latex.preamble'] = [r'\usepackage{fouriernc}']

colors = ['steelblue', 'darksalmon', 'cadetblue', 'slateblue']
```

```python
x = np.linspace(-4, 4, 200)

softplus = lambda x : np.log(1+np.exp(x))
gelu = lambda x : x * sp.stats.norm.cdf(x)
silu = lambda x : x / (1 + np.exp(-x))
mish = lambda x : x * np.tanh(np.log(1+np.exp(x)))
swish_2 = lambda x : x / ((3/2)*(1 + np.exp(-x)))
swish_0_5 = lambda x : x / ((2/3)*(1 + np.exp(-x)))
swish = lambda x,b : x / (b*(1 + np.exp(-x)))
```

```python
fig = plt.figure(figsize=(16,9)) 

ax = plt.axes()

plt.setp(ax.spines.values(), color=3 * [0.5])
ax.set_facecolor(3 * [0.99])
ax.tick_params(axis='x', colors=3 * [0.3])
ax.tick_params(axis='y', colors=3 * [0.3])

ax.plot(x, mish(x), label='Mish', color='coral')
ax.plot(x, silu(x), label='SiLU', color='seagreen')
ax.plot(10, 10, label='Swish', color='steelblue', linestyle='--')

N = 5
for i in range(1,N):
    ax.plot(x, swish(x, 1 + i / N), color='steelblue', alpha=0.85 * (1-i/(N)), linestyle='--')
    ax.plot(x, swish(x, 1 - i / N), color='steelblue', alpha=0.85 * (1-i/(N)), linestyle='--')
    
ax.set_xlim(x[0]+0.5,x[-1]-0.5)
ax.set_ylim(-0.7,4)

ax.axvline(x=0, color='black', linestyle=':', alpha=0.5)
ax.axhline(y=0, color='black', linestyle=':', alpha=0.5)

ax.legend()
ax.set_axisbelow(True)
ax.grid(color=3 * [0.88])

plt.savefig('mish_silu_swish.pdf')
```
