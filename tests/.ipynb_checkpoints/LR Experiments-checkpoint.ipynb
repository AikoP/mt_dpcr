```python
N = [25,50,75,100]
L = [None] * len(N)
M = [None] * len(N)

Nx = lambda x, mu, sigma_2 : np.exp(-np.square(x-mu)/(2*sigma_2))

for (i,n) in enumerate(N):
    
    sigma_2 = np.square(n / 4)
    #lam = lambda epoch: 10*(1-np.exp(-np.square(epoch-n)/(2*sigma_2))/n)
    #lam = lambda epoch: (1-np.exp(-np.square(3*epoch/n - 3)) * (15/n))
    
    #lam = lambda epoch: 1+(Nx(epoch, n, sigma_2) - Nx(epoch+1, n, sigma_2))/(1-Nx(epoch, n, sigma_2))
    
    lam = lambda epoch: (1-Nx(epoch + 1, n, sigma_2)) / (1-Nx(epoch, n, sigma_2))
    #lam = lambda epoch: np.exp(-(np.square(epoch+1-n)+np.square(epoch-n))/(2*sigma_2))

    L[i] = 1.0 * np.ones(n)
    M[i] = np.ones(n)
    
    for j in range(1,n):
        M[i][j] = lam(j)
        #print (M[i][j])
        L[i][j] = L[i][j-1] * M[i][j]
        if (n==100 and j==30):
            L[i][j] *= 0.8
        if (n==100 and j==70):
            L[i][j] *= 1.2
        #print (L[i][j])
    
    
#print (lr_arr)

from matplotlib import rcParams
rcParams['font.family'] = 'Century Schoolbook'
rcParams['font.cursive'] = 'Century Schoolbook'
rcParams['font.size'] = '19'

rcParams['text.usetex'] = True 
rcParams['text.latex.preamble'] = [r'\usepackage{fouriernc}']

colors = ['steelblue', 'darksalmon', 'cadetblue', 'slateblue']

fig = plt.figure(figsize=(16,9))

ax = plt.axes()

plt.setp(ax.spines.values(), color=3 * [0.5])
ax.set_facecolor(3 * [0.99])
ax.tick_params(axis='x', colors=3 * [0.3])
ax.tick_params(axis='y', colors=3 * [0.3])
    
for (i,n) in enumerate(N):
    c = colors[i]
    ax.plot(1 + np.arange(n), L[i], label=r'$\ell_t, T=%d$' % (n), color=c)
    ax.plot(1 + np.arange(n), M[i], label=r'$m_t, T=%d$' % (n), color=c, linestyle=':')
    
    
#for i in range(4):
#    plt.axhline(y = 0.2 + i * 0.2, color='grey', alpha=0.1)
    
#ax.axhline(y = 0, color='black', alpha=0.3,linewidth=1.0)
#ax.axhline(y = 1, color='black', alpha=0.3,linewidth=1.0)

ax.fill_between(-5 + np.arange(N[-1] + 11), 0, -1, alpha=0.1, color='grey')
ax.fill_between(-5 + np.arange(N[-1] + 11), 1, 2, alpha=0.1, color='grey')
    
ax.set_ylim(-0.05,1.05)
ax.set_xlim(-5,N[-1]+5)
    ax.set_xlabel(r'Number of Epochs $t$', labelpad=20)
ax.legend()
ax.set_axisbelow(True)
ax.grid(color=3 * [0.88])

plt.savefig('multiplicative_lr.pdf')
```

def getBatchLoss(model, pts, batch_ids, neighbors_dirs, perms, neighbors_dirs_perms, loss_function):

    k = neighbors_dirs.size(1)
    device = pts.device

    input_tensor = pts[batch_ids].transpose(1,2).contiguous()   # size: (batchsize, 3, N_batch)

    batch_prediction = model(input_tensor)                              # size: (batchsize, 3 * NC, N_batch)
    batch_prediction = batch_prediction.transpose(1,2).contiguous()     # size: (batchsize, N_batch, 3 * NC)

    # accumulate loss over batch ---------------------------------

    batch_loss = 0.0

    for i in range(batch_prediction.size(0)):

        # isolate sample prediction (p) and obtain all possible neighbor prediction (t_perm)
        p = batch_prediction[i].squeeze(0).reshape((-1, k, 3))              # size: (N, NC, 3)
        # t_perm = neighbors_dirs_perms[batch_pts_ids[i].flatten()]         # size: (N, perm_count, NC, 3)
        t_perm = neighbors_dirs_perms[batch_ids[i].flatten()].to(device)    # size: (N, perm_count, NC, 3)

        # for each vertex compute a permutation of neighbors s.t. the sum of differences to the prediction is minimized
        minmatches = utils.matchPoints(p.detach(), y=None, perms=perms, y_perms=t_perm)  # size: (N, NC)

        # expand the permutation index
        gather_idx = minmatches.expand(3,minmatches.size(0),minmatches.size(1)).transpose(0,1).transpose(1,2)  # size: (N, NC, 3)

        # gather neighbor directions according to the permutations
        t = neighbors_dirs[batch_ids[i].flatten()].gather(1, gather_idx)    # size: (N, NC, 3)

        batch_loss = batch_loss + loss_function(p, t)

    return batch_loss

```python
def getBatchLoss(model, pts, batch_ids, neighbors_dirs, perms, neighbors_dirs_perms, loss_function):
    
    mse_loss = torch.nn.MSELoss(reduction='mean')

    k = neighbors_dirs.size(1)
    device = pts.device

    input_tensor = pts[batch_ids].transpose(1,2).contiguous()   # size: (batchsize, d, N_batch)
    
    batchsize = input_tensor.size(0)
    d = input_tensor.size(1)
    N_batch = input_tensor.size(2)

    y = model(input_tensor)                              # size: (batchsize, d * k, N_batch)
    y = y.transpose(1,2).contiguous()                    # size: (batchsize, N_batch, d * k)
    y = y.view((batchsize, N_batch, -1, d))              # size: (batchsize, N_batch, k, d)
    
    y_norm = torch.sqrt(torch.sum(torch.square(y), dim=-1)) # size: (batchsize, N_batch, k)
    y_n = y / y_norm.unsqueeze(-1)   # size: (batchsize, N_batch, k, d)

    yy = 2 * torch.matmul(y_n, y_n.transpose(-2, -1))     # (batchsize, N_batch, k, k)
    y2 = torch.sum(y_n**2, dim=-1, keepdim=True)       # (batchsize, N_batch, k, 1)

    # internal directions
    dir_yy = y_n.unsqueeze(-2) - y_n.unsqueeze(-3) # (batchsize, N_batch, k, k, d)

    # calculate spring force by using inner distances (of normed input vectors)
    Sf_yy = torch.exp(-(y2 - yy + y2.transpose(-2, -1))) # (batchsize, N_batch, k, k)

    # calculate displacement
    Dp_yy = torch.sum(dir_yy * Sf_yy.unsqueeze(-1), dim=-2).squeeze(-2) # (batchsize, N_batch, k, d)

    # calculate target and norm to same length as y
    #T_y = y_n + Dp_yy    # (batchsize, N_batch, k, d)
    #T_y = T_y / torch.sqrt(torch.sum(torch.square(T_y), dim=-1)).unsqueeze(-1)   # (batchsize, N_batch, k, d)
    #T_y = T_y * y_norm.unsqueeze(-1)     # (batchsize, N_batch, k, d)
    
    # calculate internal loss
    #L_yy = mse_loss(y, T_y)
    
    # add displacements to the gradients
    y.grad -= Dp_yy
    
    # ---------------------------------------------------------------------------
    
    t = neighbors_dirs[batch_ids] # (batch_size, N_batch, k, d)
    
    yt = 2 * torch.matmul(y, t.transpose(3, 2))     # (batch_size, N_batch, k, k)
    y2 = torch.sum(y**2, dim=3, keepdim=True)       # (batch_size, N_batch, 1, k)
    t2 = torch.sum(t**2, dim=3, keepdim=True)       # (batch_size, N_batch, 1, k)
    
    # distances to solutions
    D_yt = y2 - yt + t2.transpose(3, 2) # (batch_size, N_batch, k, k)
    D_yt[D_yt == 0] = torch.max(D_yt) + 1.0
    
    # minimum distances to each solution
    Dt_min = torch.min(D_yt, dim = 2)[1]      # (batch_size, N_batch, k)
    #print ("Dt_min:", Dt_min)
    #print ("Dt_min.size():", Dt_min.size())
    #print ("T:", t[Dt_min])
    #print ("T.size():", t[Dt_min].size())
    
    t_min = t.gather(dim=2, index=Dt_min.unsqueeze(-1).expand((-1,-1,-1,d))) # (batch_size, N_batch, k, d)
    #print ("t_min:", t_min)
    #print ("t_min.size():", t_min.size())
    
    L_yt = mse_loss(y, t_min)
    
    # sum and normalize by number of points and batch size
    #Dt_min_sum = torch.sum(Dt_min) / (NC * N_batch * batchsize)
    
    #print ("Internal Loss:   %.20f" % (L_yy))
    #print ("Prediction Loss: %.20f" % (L_yt))
    
    #batch_loss = torch.exp(-torch.sum(D_yy) / (NC ** 2)) + torch.tanh(torch.sum(Dt_min) / NC)
    #batch_loss = L_yy + L_yt

    return L_yy, L_yt
```

```python
torch.cuda.empty_cache()

mini_batch_size = 1
loss_function = torch.nn.MSELoss(reduction='mean')
lr = 0.003

class AnnealingFactor(object):
    def __init__(self, epochs=20):
        self.epochs = epochs
    def __call__(self, epoch):
        return -np.exp(-np.square((epoch-self.epochs)/(self.epochs*0.3)))/(self.epochs/20) + 1

model = predictor_bn.Model(d = 3, k = 20, emb_dims = 1024,neighbors = 6).to(device)
model = torch.nn.DataParallel(model)

#checkpoint = torch.load('../training/checkpoints/predictor_bn_model/2020-10-03_2059/predictor_bn_model.t7')
#model.load_state_dict(checkpoint['model_state_dict'][-1])

model.train()

opt = Ranger(model.parameters(), lr=lr)
scheduler = MultiplicativeLR(opt, lr_lambda=AnnealingFactor(20))

binbatches = utils.drawBinBatches([bin.size(0) for bin in train_bins], batchsize=mini_batch_size)

k = 0

start = time.time()

for (bin_id, batch_ids) in binbatches:
    
    if k>100:
        break

    # prediction & loss ----------------------------------------
    

    batch_pts_ids = train_bins[bin_id][batch_ids][:,:,0]    # size: (batchsize, N_batch)
    L_yy, L_yt = getBatchLoss(model, pts, batch_pts_ids, neighbors_dirs, perms, neighbors_dirs_perms, loss_function)
        
    #print ("Internal Loss:   %.20f" % (L_yy))
    #print ("Prediction Loss: %.20f" % (L_yt))
    
    #with torch.no_grad():
    #    old_loss = getOldBatchLoss(model, pts, batch_pts_ids, neighbors_dirs, perms, neighbors_dirs_perms, loss_function) / mini_batch_size
    #print ("Old Loss:        %.20f" % (old_loss))
    
    batch_loss = L_yy + L_yt
    batch_loss.backward()
    
    lr_adjust = 1.0
    
    model_state_backup = copy.deepcopy(model.state_dict())
    opt_state_backup = copy.deepcopy(opt.state_dict())
    
    opt.step()
    
    new_batch_loss = 0.0
    
    for i in range(10):
        
        with torch.no_grad():
            L_yy, L_yt = getBatchLoss(model, pts, batch_pts_ids, neighbors_dirs, perms, neighbors_dirs_perms, loss_function)
            new_batch_loss = L_yy + L_yt
        
        if (new_batch_loss.item() < batch_loss.item()):
            break
        else:
            model.load_state_dict(model_state_backup)
            opt.load_state_dict(opt_state_backup)
            lr_adjust /= 2.0
            opt.step(lr_adjust = lr_adjust)
        
    if new_batch_loss.item() > batch_loss.item():
        
        print ("Batch %2d loss: %.20f -> failed to optimize (%.19f) [%3d]" % (k, batch_loss.item(),new_batch_loss.item(), time.time() - start))
        
    else:
        
        print ("Batch %2d loss: %.20f -> Min @ %.9f : %.20f [%3d]" % (k, batch_loss.item(),lr_adjust,new_batch_loss.item(), time.time() - start))
    
    k += 1
```
