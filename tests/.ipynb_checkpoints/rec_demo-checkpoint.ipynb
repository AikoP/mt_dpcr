```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

import numpy as np
import torch

from utils import reader
from utils import dpcrUtils as utils
from utils import generator

import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 50
a = 0.5

x = np.linspace(-a,a,N)
y = np.linspace(-a,a,N)

X = np.concatenate((np.repeat(x, N)[:, None], np.tile(x, N)[:, None]), axis=1)
```

```python
Z = np.sum(np.square(X), axis = 1)

fig = plt.figure(figsize=(16,16))
ax = Axes3D(fig)

ax.scatter(X[:,0], X[:,1], Z[:])

plt.show()
```

```python
device = utils.getDevice()

pts_np = np.concatenate((X,Z[:,None]), axis=1)

model = torch.from_numpy(pts_np).to(device)
knn = utils.knn(model, 6)
```
