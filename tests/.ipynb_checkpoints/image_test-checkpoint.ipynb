```python
import sys
sys.path.insert(1, '../')  # to load from any submodule in the repo

from PIL import Image
import numpy as np
import torch
import matplotlib
import matplotlib.pyplot as plt

import utils.dpcr_generator as generator
```

```python
img = Image.open('A.png') # use pillow to open a file
# img = img.resize((self.width, self.height)) # resize the file to 256x256
img = img.convert('RGB') #convert image to RGB channel
img = (255 - np.asarray(img).max(axis = -1)) / 255.0 # we have to change the dimensions from width x height x channel (WHC) to channel x width x height (CWH)
img = torch.from_numpy(np.asarray(img)) # create the image tensor

print (torch.sum(img))
```

```python
Nx = img.size(0)
Ny = img.size(1)

X = torch.arange(Nx).repeat(Ny,1)
Y = torch.arange(Ny).view((-1,1)).repeat(1,Nx)

XY = torch.zeros((Nx,Ny,2))
XY[:,:,0] = X
XY[:,:,1] = Y

pts =  XY[img==1]
pts[:,1] *= -1

pts += 0.15 * torch.randn(pts.size())
```

```python
plt.figure(figsize=(9,9))
plt.scatter(pts.numpy()[:,0], pts.numpy()[:,1])
plt.axis('equal')
plt.show()
```

```python
samples = generator.getData(pts, 1, noise = 0, hidden_range = (3, 16), iteration_range = (3,6))
```

```python
idx = np.random.randint(len(samples))

sample_pts = pts[samples[idx][:,0]]
edge_mask = samples[idx][:,1]
sample_pts_np = sample_pts.numpy()

plt.figure(figsize=(9,9))
plt.scatter(sample_pts_np[edge_mask == 0,0], sample_pts_np[edge_mask == 0,1])
plt.scatter(sample_pts_np[edge_mask == 1,0], sample_pts_np[edge_mask == 1,1])
plt.axis('equal')
plt.ylim(-50,0)
plt.show()
```
