```python
def knn(x, k):

    """
        x: (b x d x n) tensor
        k: integer
    """

    # compute pairwise distances
    inner = torch.matmul(x.transpose(2, 1), 2.0 * x)        # (b x n x n)
    xx = torch.sum(torch.square(x), dim=1, keepdim=True)    # (b x 1 x n)
    pairwise_distance = xx - inner + xx.transpose(2, 1)     # (b x n x n)
 
    return pairwise_distance.topk(k=k+1, dim=-1, largest=False)[1][:,:,1:]   # (b x n x k)
    #return pairwise_distance.topk(k=k, dim=-1, largest=False)[1]   # (b x n x k)


def get_graph_feature(x, k = 20):

    """
        x: (b x d x n) tensor
        k: integer 
    """
    
    #print (x)

    b, d, n = x.size()

    idx_base = torch.arange(b, device=x.device).view(-1, 1, 1) * n   # (b x 1 x 1)

    idx = (knn(x, k = k) + idx_base).view(-1)   # (b x n x k) -> (b * n * k)

    x = x.transpose(2, 1).contiguous()          # (b, d, n)  -> (b, n, d)
    feature = x.view(b*n, -1)[idx, :]           # (b, n, d)  -> (b * n, d) -> (b * n * k, d)
    feature = feature.view(b, n, k, d)          # (b * n * k, d) -> (b, n, k, d)
    x = x.view(b, n, 1, d).repeat(1, 1, k, 1)   # (b, n, d) -> (b, n, 1, d) -> (b, n, k, d)
    
    feature = torch.cat((feature-x, x), dim=3).permute(0, 3, 1, 2)  # (b, n, k, 2 * d) -> (b, 2 * d, n, k)
    
    return feature      # (b, 2 * d, n, k)

def get_new_graph_feature(x, k = 20):

    """
        x: (b x d x n) tensor
        k: integer 
    """

    b, d, n = x.size()
    
    knn_idx = knn(x, k = k).unsqueeze(1).expand(-1, d, -1, -1)   # (b x n x k) -> (b x 1 x n x k) -> (b x d x n x k)

    f = x.unsqueeze(-2).expand(-1, -1, n, -1).gather(-1, knn_idx)   # (b x d x n) -> (b x d x n x 1) -> (b x d x n x n) -> (b x d x n x k)
    x = x.unsqueeze(-1).expand(-1, -1, -1, k)                       # (b x d x n) -> (b x d x n x 1) -> (b x d x n x k)
    
    return torch.cat([f-x, x], dim=1)  # (b x 2*d x n x k)
```
